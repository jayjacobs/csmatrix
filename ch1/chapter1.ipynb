```python
# Task 1.4.10
from image import file2image

# this does not return intensities.
data = file2image("../data/img01.png")

height = len(data)-1
width = len(data[0])-1

# use sum() to approximate intensities maybe?
pts  = [ x + (len(data)-1-y)*1j for y in range(height) for x in range(width) if sum(data[y][x]) < 400 ]

HTML(plot(pts, scale=height, dot_size=1))  
```

```python
# Task 1.4.11
# adjust it by half of width and half of height
t_pts = [ x-((width/2)+(height/2)*1j) for x in pts ]

# make scale half as much for fun
HTML(plot(t_pts, scale=height/2, dot_size=2))  
```

```python
# Task 1.4.12
tr_pts = [ (-0.5j*z) for z in t_pts ]

# make scale half as much for fun (but points are 1/2'd anyway)
HTML(plot(tr_pts, scale=height/2, dot_size=2))  
```
